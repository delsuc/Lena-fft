# Lena FFT

This little example show how you can Fourier transform an image and display both amplitude and phase.

Note how the phase contains much more information than the amplitude.

This little code is to be compared with the same exercise in Mathematica, found here : <br/>
http://mathematica.stackexchange.com/questions/29203/determine-the-2d-fourier-transform-of-an-image/29204#29204

```python
%pylab inline
from scipy import misc, fftpack      # misc contains the lena picture
image = misc.lena().astype(float32)  # that is imported a float 2D array
plt.gray()                           # display in gray
imshow(image)
```

```python
# fftshift swaps the result so that zero frequency is in the center
spectrum = fftpack.fftshift(fftpack.fft2(image))
```

```python
imshow(log(1+abs(spectrum)))
title("magnitude spectrum")
```

```python
phase = arctan2(spectrum.imag, spectrum.real)
imshow(phase)
title("phase spectrum")
```
